```python
# prompt: Identify the entities and relationships in a given file using NLP and output in nlp.txt

!pip install spacy
!python -m spacy download en_core_web_sm

import spacy
nlp = spacy.load("en_core_web_sm")

def extract_entities_relationships(file_path):
    try:
        with open(file_path, 'r') as file:
            text = file.read()
    except FileNotFoundError:
        return "File not found."

    doc = nlp(text)
    entities_relationships = []
    for token in doc:
        if token.dep_ in ("nsubj", "dobj", "pobj", "attr"): # Common dependency relations
          entities_relationships.append((token.text, token.dep_, token.head.text))

    return entities_relationships

def write_to_file(output_path, content):
    with open(output_path, "w") as f:
      for entity_rel in content:
          f.write(f"{entity_rel[0]} ({entity_rel[1]}) {entity_rel[2]}\n")

# Example Usage:
file_path = "/content/drive/MyDrive/EOY-report.txt" # Replace with the actual path to your file
output_file = "nlp.txt"
results = extract_entities_relationships(file_path)
if isinstance(results, str):
    print(results) # Handle file not found error
else:
    write_to_file(output_file, results)
    print(f"Entities and relationships written to {output_file}")
```
